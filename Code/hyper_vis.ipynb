```python
# create visualizations of holomorphic functions

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import axes3d
import cmath as cm


# create a grid of complex numbers
def create_grid(xmin, xmax, ymin, ymax, nx, ny):
    x = np.linspace(xmin, xmax, nx)
    y = np.linspace(ymin, ymax, ny)
    return np.meshgrid(x, y)

# plot a complex function
def plot_complex_function(f, xmin, xmax, ymin, ymax, nx, ny, title):
    X, Y = create_grid(xmin, xmax, ymin, ymax, nx, ny)
    Z = X + 1j*Y
    W = [[f(z) for z in row] for row in Z]
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, np.real(W), cmap='viridis', norm=colors.Normalize(vmin=-1, vmax=1))
    ax.set_xlabel('$\\mathrm{Re}(z)$')
    ax.set_ylabel('$\\mathrm{Im}(z)$')
    ax.set_title(title)
    plt.show()

tau = -1/(2 * np.pi * 1j)
H = lambda z: tau*cm.log(-z)
delta = lambda z: tau/z

```

```python
plot_complex_function(H, -2.5, 3.5, -1, 1, 500, 500, 
                      "$\\mathrm{Re}\\left(\\frac{-\\log(-z)}{2\\pi i}\\right)$")

```

```python
plot_complex_function(delta, -1, 1, -0.5, 0.5, 200, 200,"$\\mathrm{Re}\\left(\\frac{-1}{2\\pi i z}\\right)$")
```
